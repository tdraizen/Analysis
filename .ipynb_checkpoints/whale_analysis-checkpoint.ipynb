# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:
1. whale_returns.csv
2. algo_returns.csv
3. sp500_history.csv

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns_path = Path("whale_returns.csv")
whale_returns_df = pd.read_csv(whale_returns_path, index_col='Date', infer_datetime_format=True, parse_dates=True)
whale_returns_df.head()
```

```python
# Count nulls
whale_returns_df.isnull().sum()
```

```python
# Drop nulls
whale_returns_df.dropna(inplace=True)
whale_returns_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_path = Path("algo_returns.csv")
algo_returns_df = pd.read_csv(algo_returns_path, index_col='Date', infer_datetime_format=True, parse_dates=True)
algo_returns_df.head()
algo_returns_df.dtypes
```

```python
# Count nulls
algo_returns_df.isnull().sum()
```

```python
# Drop nulls
algo_returns_df.dropna(inplace=True)
algo_returns_df.isnull().sum()
algo_returns_df.head()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices, sorting index
sp500_historical_path = Path("SP500_historical.csv")
sp500_historical_df = pd.read_csv(sp500_historical_path, index_col='Date', infer_datetime_format=True, parse_dates=True)
sp500_historical_df.head()
```

```python
# Check Data Types
sp500_historical_df.dtypes
sp500_historical_df.isnull().sum()
```

```python
# Fix Data Types
# Take out the $ data so you can do calculations
sp500_historical_df["Close"] = sp500_historical_df["Close"].str.replace("$", "")
sp500_historical_df["Close"] = sp500_historical_df["Close"].str.replace(",", "")
sp500_historical_df["Close"] = sp500_historical_df["Close"].astype("float")
sp500_historical_df = sp500_historical_df.astype(float)
sp500_historical_df.head()
sp500_historical_df.dtypes
```

```python
# Calculate Daily Returns
sp500_daily_returns_df = sp500_historical_df.pct_change()
sp500_daily_returns_df.head()
```

```python
# Drop nulls
sp500_daily_returns_df.dropna(inplace=True)
sp500_daily_returns_df.head()
```

```python
# Rename Column
sp500_daily_returns_df.columns = ['SP500']
sp500_daily_returns_df.dtypes
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
combined_ports_df = pd.concat([sp500_daily_returns_df, whale_returns_df, algo_returns_df], axis ="columns", join="inner")
combined_ports_df.head()
```

### Challenge: Exponentially Weighted Average 

An alternative way to calculate a rollwing window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the `ewm` with a 21 day half-life.

```python
expo_moving_21 = combined_ports_df.ewm(span = 21, min_periods=21,adjust=False).mean()
expo_moving_21.plot(figsize= (20,10), title='21 Day Exponential Moving Average')
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

Calculate and plot the annualized Sharpe ratios for all portfolios to determine which portfolio has the best performance

```python
# Annualzied Sharpe Ratios
sharpe_ratios_combined_ports = (combined_ports_df.mean() * 252) / (combined_ports_df.std() * np.sqrt(252))
sharpe_ratios_combined_ports.sort_values(ascending=False)
```

 plot() these sharpe ratios using a barplot.
 On the basis of this performance metric, do our algo strategies outperform both 'the market' and the whales?

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios_combined_ports.plot(kind='bar',figsize=(18,9), title= 'Combined Portolio Sharpe Ratios')
```

```python
# Read the first stock
EEM_path = Path("EEM.csv")
EEM_df = pd.read_csv(EEM_path, index_col='Date', infer_datetime_format=True, parse_dates=True)
EEM_df.columns = ['EEM']
EEM_df.head()
```

```python
# Read the second stock
JPM_path = Path("JPM.csv")
JPM_df = pd.read_csv(JPM_path, index_col='Date', infer_datetime_format=True, parse_dates=True)
JPM_df.columns = ['JPM']
JPM_df.head()
```

```python
# Read the third stock
LMT_path = Path("LMT.csv")
LMT_df = pd.read_csv(LMT_path, index_col='Date', infer_datetime_format=True, parse_dates=True)
LMT_df.drop(columns=['Unnamed: 2'], inplace =True)
LMT_df.columns = ['LMT']
LMT_df.head()
```

```python
# Read the fourth stock
PTY_path = Path("PTY.csv")
PTY_df = pd.read_csv(PTY_path, index_col='Date', infer_datetime_format=True, parse_dates=True)
PTY_df.columns = ['PTY']
PTY_df.head()
```

```python
# Read the fifth stock
SGEN_path = Path("SGEN.csv")
SGEN_df = pd.read_csv(SGEN_path, index_col='Date', infer_datetime_format=True, parse_dates=True)
SGEN_df.columns = ['SGEN']
SGEN_df.head()
```

```python
# Concatenate all stocks into a single DataFrame
troy_portfolio = pd.concat([EEM_df, JPM_df, LMT_df, PTY_df, SGEN_df], axis='columns', join='inner')
#troy_portfolio.isnull().sum()
#troy_portfolio.head()
troy_daily_returns_df = troy_portfolio.pct_change()
troy_daily_returns_df.head()
```

```python
# Reset the index
# Set Index on Previous read in's
```

```python
# Drop Nulls
troy_daily_returns_df.isnull().sum()
troy_daily_returns_df.dropna(inplace=True)
troy_daily_returns_df.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/5, 1/5, 1/5, 1/5, 1/5]
troy_weighted_port_daily_df = troy_daily_returns_df.dot(weights)
troy_weighted_port_daily_df.head()
```
